```python
%%writefile /content/data.yaml
# Soil Detection Dataset configuration for YOLOv11

# Root dataset directory
path: /content/drive/MyDrive/Soil detection.v3i.yolov11

# Training, validation, and testing image directories
train: /content/drive/MyDrive/Soil detection.v3i.yolov11/train/images
val: /content/drive/MyDrive/Soil detection.v3i.yolov11/valid/images
test: /content/drive/MyDrive/Soil detection.v3i.yolov11/test/images

# Class names (must match your labeling order)
names:
  0: Alluvial Soil
  1: Black Soil
  2: Clay Soil
  3: Red Soil
```

```python
from ultralytics import YOLO

# Load a pretrained YOLOv11 model (use yolo11n.pt or yolo11s.pt depending on GPU)
model = YOLO('yolo11n.pt')  # small & fast; for better accuracy try 'yolo11s.pt'

# Start training
model.train(
    data='/content/data.yaml',
    epochs=100,
    imgsz=640,
    batch=16,
    project='/content/drive/MyDrive/yolov11_runs',
    name='soil_detection_exp',
    workers=8,
    patience=20
)
```

```python
from ultralytics import YOLO

# Load best weights
model = YOLO('/content/drive/MyDrive/yolov11_runs/soil_detection_exp3/weights/best.pt')

# Run prediction on a folder of test images
model.predict(
    source='/content/drive/MyDrive/Soil detection.v3i.yolov11/test/images',
    imgsz=640,
    conf=0.25,
    save=True,
    project='/content/drive/MyDrive/yolov11_runs',
    name='soil_detection_predictions'
)
```

```python
# Evaluate performance on validation set
metrics = model.val(data='/content/data.yaml')
print("mAP50-95:", metrics.box.map)
print("Precision:", metrics.box.mp)
print("Recall:", metrics.box.mr)
```
